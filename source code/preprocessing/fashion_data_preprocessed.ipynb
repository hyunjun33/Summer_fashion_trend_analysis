```python
from nltk.util import ngrams
```

```python
def ngram_vector(x):
    remove_words = ["(", ")", "[", "]","★"] 
    x_list = x.split(" ")
    ngram_list = []
    for word in x_list:
        for j in range(len(remove_words)):
            word = word.replace(remove_words[j], "")
        ngram_list.append(word)
    
    return ngram_list 
```

```python
def listed_item(x):
    item_list = []
    for item in x:
        n_gram_final = " ".join(list(item))
        item_list.append(n_gram_final)
    return item_list
```
