```python
after_stop_word = []

for hash_tag in hash_tag_column:
    if hash_tag not in hash_tag_stop_words:
        after_stop_word.append(hash_tag)

after_stop_word = nltk.Text(after_stop_word, name="해시태그")
plt.figure(figsize=(12, 6))
after_stop_word.plot(50)
```

```python
import platform
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
```

```python
hashtag_mask = np.array(Image.open("mobile_phone_icon.jpg"))
data = after_stop_word.vocab().most_common(50)
```

```python
wc = WordCloud(background_color="white",
               max_words=200,
               mask=hashtag_mask,
               contour_width=3,
               contour_color="steelblue",
               font_path=font_path).generate_from_frequencies(dict(data))
```

```python
plt.figure(figsize=(12,8))
plt.imshow(wc)
plt.axis("off")
plt.show()
```
