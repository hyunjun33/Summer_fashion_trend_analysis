```python
data = pd.read_csv('new_2021_fashion_news.csv', encoding='utf-8')
with open('new_2021_fashion_news.txt','w', encoding='utf-8') as f:
    for line in data.values:
        f.write((str(line[0])+'\t'+str(line[1])+'\n'))
os.listdir( os.getcwd() )
```

```python
doc_ko = open("new_2021_fashion_news.txt").read() #용량이 너무크면 불러오지 못알 수 있음, 그렇다면 readline으로 한줄씩 불러와야함
doc_ko[1:1000]
```

```python
from konlpy.tag import Okt
t = Okt()

doc_nouns = t.nouns(doc_ko)
print( len( doc_ko))
len(doc_nouns)
```

```python
# nltk.Text()를 이용하여 nltk가 가지는 많은 기능을 사용 가능해짐.
ko = nltk.Text(doc_nouns, name="패션")
print(type(ko))
print(len(ko.tokens))
```

```python
### 중복된 단어를 제거한 개수를 확인
print(len(set(ko.tokens)))
```

```python
ko.vocab().most_common(50)
```

```python
ko = nltk.Text(ko, name="패션")
plt.figure(figsize=(16,9))
ko.plot(50)
```

```python
f=open("stopwords_fashion.txt")
lines = f.readlines()

stop_words = []
for one in lines:
  stop_words.append(one.strip())


new_ko = [ ]
for one_word in ko:
  if one_word not in stop_words:
    new_ko.append(one_word)
```

```python
new_ko = nltk.Text(new_ko, name="패션")
plt.figure(figsize=(16,9))
new_ko.plot(50)
```

```python
new_ko.vocab().most_common(50)
```
