```python
import sys
import io
import os

from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
import urllib.parse
from urllib.request import Request, urlopen
from time import sleep
import pandas as pd
```

```python
search = input("검색어를 입력하세요: ")
searching =str(search)

search = urllib.parse.quote(search)
```

```python
url = "https://www.instagram.com/explore/tags/{}/".format(search)
driver = webdriver.Chrome('./chromedriver_90')

driver.get(url)
sleep(5)

SCROLL_PAUSE_TIME = 1.2 # 인스타게시물 스크롤 속도 조절
```

```python
reallink = []

while True:
    page_string = driver.page_source
    soup = BeautifulSoup(page_string, "lxml")
    
    for link1 in soup.find_all(name="div", attrs={"class":"Nnq7C weEfm"}):
        title = link1.select("a")[0]
        real = title.attrs["href"]
        reallink.append(real)
        
        title = link1.select("a")[1]
        real = title.attrs["href"]
        reallink.append(real)
        
    last_height = driver.execute_script("return document.body.scrollHeight")
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        sleep(SCROLL_PAUSE_TIME)
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        
        else:
            last_height = new_height
            continue
```
