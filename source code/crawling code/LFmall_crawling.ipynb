```python
from selenium import webdriver
driver = webdriver.Chrome('chromedriver_90')
driver
```

```python
url = 'https://www.lfmall.co.kr/p2/display/menu?id=501'
driver.get(url)
```

```python
link_obj = driver.find_element_by_xpath('//*[@id="menu-button-502"]/a')
link_obj.click()
```

```python
from IPython.display import display, Image
import os, warnings
import re
warnings.filterwarnings(action='ignore')
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
# 여성복 베스트 100
link_obj = driver.find_element_by_xpath('/html/body/div/div[4]/div/div[1]/div[2]/div/div[1]/ul/li[2]/a/span/img')
link_obj.click()
```

```python
import time
for i in range(0,3):
    time.sleep(1)
    driver.execute_script("window.scrollTo(0, 200)")
```

```python
best_item = driver.find_elements_by_class_name('prd-desc')
print(len(best_item))
```

```python
all_brand = []
all_name = []
all_price = []
all_review = []

for product in best_item:
    brand = product.find_element_by_class_name('brand').text
    name = product.find_element_by_class_name('name').text
    price = product.find_element_by_class_name('discount').text
    review = product.find_element_by_class_name('point').text
    
    all_brand.append(brand)
    all_name.append(name)
    all_price.append(price)
    all_review.append(review)
    
print(len(all_brand))
print(len(all_name))
print(len(all_price))
print(len(all_review))
print(all_brand)
print(all_name)
print(all_price)
print(all_review)
```

```python
LFmall = {}

LFmall["brand"] = all_brand
LFmall["name"] = all_name
LFmall["price"] = all_price
LFmall["review"] = all_review

LFmall
```

```python
import pandas as pd

LFmall_df_female = pd.DataFrame(LFmall)
LFmall_df_female.to_excel("LFmall_female.xlsx")
```
